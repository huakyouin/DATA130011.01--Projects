将tensorboard上的曲线数据下载下来手动可视化，添加图例，横纵坐标等信息

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
import pylab
import sys

pylab.rcParams['figure.figsize'] = (10.0, 4.0)
colors=['b','g','r','c','m']
filenames=["use erase_logs_within Acc_train","use erase_logs_within Acc_test"]
labels=['train Acc','test Acc']
for item,color,label in zip(filenames,colors,labels):
    pn=pd.read_csv(sys.path[0]+'/'+item+'.csv', usecols=['Step', 'Value'])
    plt.plot(pn.Step, pn.Value, lw=1.5, label=label, color=color)

plt.xlabel('epoch')
# plt.xticks(range(0,20,2))
plt.ylabel('Accuary')
# plt.ylabel('loss')
plt.title('Use Erase')
# plt.ylabel('learning rate')
plt.legend(loc='best')
plt.show()

# tensorboard --logdir=tensorboardData
```
